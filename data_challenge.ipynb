# Data Preprocessing
Much of the data is categorical. Many machine learning algorithms require numerical data. The data we are working with is nominal data, so using the one-hot encoding technique makes the most 
sense for the encoding strategy.

```python
from sklearn.preprocessing import OneHotEncoder
label_encoder = OneHotEncoder(sparse_output=False)
train_df_cat_enc = label_encoder.fit_transform(train_df[['Operating Airline', 'Published Airline', 'GEO Summary', 'GEO Region', 'Activity Type Code', 'Cargo Type Code',
       'Cargo Aircraft Type']])
train_df_enc = pd.DataFrame(train_df_cat_enc)
train_df_enc['Activity Period'] = train_df['Activity Period']
train_df_enc['Cargo Metric TONS'] = train_df['Cargo Metric TONS']
train_df_enc
```

```python
test_df_cat_enc = label_encoder.fit_transform(test_df[['Operating Airline', 'Published Airline', 'GEO Summary', 'GEO Region', 'Activity Type Code', 'Cargo Type Code',
       'Cargo Aircraft Type']])
test_df_enc = pd.DataFrame(test_df_cat_enc)
test_df_enc['Activity Period'] = test_df['Activity Period']
test_df_enc
```

```python
encoded_train_df = pd.get_dummies(train_df)
encoded_test_df = pd.get_dummies(test_df)
encoded_train_df, encoded_test_df = encoded_train_df.align(encoded_test_df, axis=1, join='outer', fill_value=0)
```

```python
encoded_train_df
```

```python
encoded_test_df
```

We now need to split the test data further into a training set and a validation set. I will use and 80/20 split for training/test data.

```python
import numpy as np

def split_train_test(df, test_ratio):
    shuffled_indices = np.random.permutation(len(df))
    test_set_size = int(len(df) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return df.iloc[train_indices], df.iloc[test_indices]

train_set, test_set = split_train_test(encoded_train_df, 0.2)
```

# Build And Evaluate Models
We are trying to predict the cargo weight of any given shipment. Given that we know the target, and the target is a real a number, there are a few algorithms that are best suited for this problem.
I will be evaluating the performance of a linear regression model, a Ridge model, a LASSO mdoel, a basic decision tree, and XG Boost.

```python
train_y = train_set['Cargo Metric TONS']
test_y = test_set['Cargo Metric TONS']
train_x = train_set.drop(columns=['Cargo Metric TONS'])
test_x = test_set.drop(columns=['Cargo Metric TONS'])
```

```python
train_y
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(train_x, train_y)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

hyp = reg.predict(test_x)
mse = mean_squared_error(test_y, hyp)
lin_reg_error = np.sqrt(mse)
lin_reg_error
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

param = {'alpha': [0.01, 0.1, 1, 2, 3, 4, 10, 15, 20, 25, 100]}
ridge = Ridge()
best_ridge = GridSearchCV(ridge, param, scoring='neg_mean_squared_error', cv=5)
best_ridge.fit(train_x, train_y)
```

```python
mse = mean_squared_error(test_y, best_ridge.predict(test_x))
ridge_err = np.sqrt(mse)
ridge_err
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
best_lasso = GridSearchCV(lasso, param, scoring='neg_mean_squared_error', cv=5)
best_lasso.fit(train_x, train_y)
```

```python
best_lasso_err = np.sqrt(mean_squared_error(test_y, best_lasso.predict(test_x)))
best_lasso_err
```

```python
from sklearn.tree import DecisionTreeRegressor

clf = DecisionTreeRegressor()
clf.fit(train_x, train_y)
```

```python
mse = mean_squared_error(test_y, clf.predict(test_x))
dec_tree_err = np.sqrt(mse)
dec_tree_err
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(train_x, train_y)
```

```python
xgb_err = np.sqrt(mean_squared_error(test_y, xgb.predict(test_x)))
xgb_err
```

# Evaluate Models
The RMSE for each model is as follows:
1. XG Boost - 129.84121978542345
